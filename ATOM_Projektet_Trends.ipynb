#Nogle generelle rutiner i Python

```python
#Return half first=1 or second=2 half year given a datetime structure
def HalfYear(datetime):
  if datetime.date().month > 6:
    return 2
  return 1
```

```python
def GetSetAnalysisDataFrame(sourceDataFrame,inclusionSet, attribute,onlyFirst=False):
  matchedList = []
  finishList = []
  halfYearList = []
  yearList = []
  programmeList = []
  otherList = []
  for idx in sourceDataFrame.index:
    row = sourceDataFrame.loc[idx]
    itemList = str(row[attribute]).split(";")
    for item in itemList:
        if (item != '') & (item != 'nan'):
           finishList.append(row['Finish'])
           halfYearList.append(HalfYear(row['Finish']))
           yearList.append(row['Finish'].date().year)
           programmeList.append(row['Programme'])
           if  (item in inclusionSet):
                matchedList.append(item)
           else:
                matchedList.append("Other")
                otherList.append(item)
        if onlyFirst == True:
           break
  data = {
     "FinishList":finishList,
     "Programme":programmeList,
     "HalfYear":halfYearList,
     "Year":yearList,
     attribute: matchedList
      }

  return pd.DataFrame(data), otherList
```
